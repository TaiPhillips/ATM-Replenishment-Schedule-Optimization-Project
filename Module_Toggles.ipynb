```r
Sub Monday_Toggle()
    
    Call Day_Toggle("Monday")
    
End Sub
Sub Tuesday_Toggle()
    
    Call Day_Toggle("Tuesday")
    
End Sub
Sub Wednesday_Toggle()
    
    Call Day_Toggle("Wednesday")
    
End Sub
Sub Thursday_Toggle()
    
    Call Day_Toggle("Thursday")
    
End Sub
Sub Friday_Toggle()

    Call Day_Toggle("Friday")
    
End Sub
Sub Saturday_Toggle()

    Call Day_Toggle("Saturday")
    
End Sub
Sub Sunday_Toggle()

    Call Day_Toggle("Sunday")
    
End Sub

Sub Day_Toggle(day_name As String)
Dim day_index As Integer
            
    day_index = Application.WorksheetFunction.Match(day_name, Range("R_Days"), 0)
    Application.ScreenUpdating = False
    Application.Calculation = xlManual
If Cells(day_index + 5, 3).Value = "" Then
    Cells(day_index + 5, 3) = "X"

Else
    Cells(day_index + 5, 3) = ""
End If
   
    Call UpdateTable
    
    Application.ScreenUpdating = True
    Application.Calculation = xlAutomatic
    

End Sub

'outdated code, please ignore, incorporated into lower section'
Sub Update_R_Schedule_Table()
Dim tableName As String

tableName = "R_Schedule_Table"

Application.ScreenUpdating = False
Application.Calculation = xlManual
Call UpdateTable

Application.ScreenUpdating = True
Application.Calculation = xlAutomatic

End Sub

'outdated code, please ignore, incorporated into lower section'
Sub UpdateDaysofWeek()
Dim dayindex(6) As Long
Dim start_index As Integer
Dim rep_index As Integer
Dim array_index As Integer
Dim array_length2 As Integer
Dim update_index As Integer
Dim array_length As Integer
Dim start_date As Date
Dim scheduleTable As String
Dim worksheetName As String
Dim newAddress As String

scheduleTable = "R_Schedule_Table"
worksheetName = "Analyze Replenishment Schedule"

array_index = 0
rep_index = 13
start_index = 6
update_index = 16

Do While start_index < 13

    If Cells(start_index, 3) = "X" Then
        dayindex(array_index) = start_index
        start_index = start_index + 1
        array_index = array_index + 1
    Else
        start_index = start_index + 1
    End If
Loop

array_length2 = array_index
array_length = array_index
array_index = 0

Do While rep_index > 0
    If array_index >= array_length Then
        rep_index = rep_index - 1
        array_index = 0
    Else
        Cells(update_index, 2) = Cells(dayindex(array_index), 4).Value
        update_index = update_index + 1
        array_index = array_index + 1
    End If
Loop

Range(scheduleTable).Cells(1, 1).Select
    newAddress = Range(Selection, Selection.Offset(13 * array_length2 - 1, 5)).address
    With ActiveWorkbook.Names(scheduleTable)
            .RefersTo = "='" & worksheetName & "'!" & newAddress
    End With
    
rep_index = 13
array_index = 0
update_index = 16
start_date = Cells(3, 5).Value



Do While rep_index > 0
    If array_index >= array_length Then
        rep_index = rep_index - 1
        array_index = 0
        start_date = start_date + 7
        
    Else
        Cells(update_index, 3) = start_date + dayindex(array_index) - 7 + 1
        update_index = update_index + 1
        array_index = array_index + 1
    End If
Loop

End Sub

Sub UpdateTable()

Dim timeColumn As String
Dim timeDateColumn As String
Dim costColumn As String
Dim amountColumn As String
Dim timeFormula As String
Dim timeDateFormula As String
Dim costFormula As String
Dim amountFormula As String
Dim numRow As Integer
Dim numCol As Integer
Dim worksheetName As String
Dim scheduleTable As String
Dim newAddress As String
Dim dayindex(6) As Long
Dim start_index As Integer
Dim rep_index As Integer
Dim array_index As Integer
Dim array_length2 As Integer
Dim update_index As Integer
Dim array_length As Integer
Dim start_date As Long
Dim end_date As Long
Dim repDate, repDay As String


Application.ScreenUpdating = False
Application.Calculation = xlManual

start_date = Range("Start_Date").Cells(1, 1).Value
end_date = Range("End_Date").Cells(1, 1).Value

If Range("Start_Date").Cells(1, 1) = "" Or Range("End_Date").Cells(1, 1) = "" Or start_date > end_date Then
        MsgBox ("Invalid Dates Entered")
        Exit Sub
    End If

repDay = "R_Schedule_Days_of_Week"
repDate = "R_Schedule_Days"
timeColumn = "R_Schedule_Times_of_Day"
timeDateColumn = "R_Schedule_Times"
costColumn = "R_Schedule_Costs"
amountColumn = "R_Schedule_Amounts"
scheduleTable = "R_Schedule_Table"
worksheetName = "Analyze Replenishment Schedule"

timeFormula = Range(timeColumn).Cells(1, 1).Formula
timeDateFormula = Range(timeDateColumn).Cells(1, 1).Formula
costFormula = Range(costColumn).Cells(1, 1).Formula
amountFormula = Range(amountColumn).Cells(1, 1).Formula

worksheetName = "Analyze Replenishment Schedule"
numRow = Range(scheduleTable).Rows.Count

array_index = 0
rep_index = (end_date - start_date + 1) / 7
start_index = 6
update_index = 16

For i = 1 To 7

    If Range("Selected_Days").Cells(i, 1) = "X" Then
        dayindex(array_index) = start_index
        start_index = start_index + 1
        array_index = array_index + 1
    Else
        start_index = start_index + 1
    End If
Next i

array_length2 = array_index
array_length = array_index
array_index = 0

If array_length = 0 Then
        MsgBox ("Must Select One Day for Replenishment")
        Call Day_Toggle(Cells(update_index, 2))
        Exit Sub
    End If


Range(scheduleTable).ClearContents
If numRow > 1 Then
        Range(Range(scheduleTable).Rows(2), Range(scheduleTable).Rows(numRow)).Clear
    End If

x = end_date - start_date + 1
y = x / 7
Z = Application.WorksheetFunction.RoundDown(y, 0)

For j = 1 To x
    If array_index >= array_length Then
        x = x - 1
        array_index = 0
    Else
        Cells(update_index, 2) = Cells(dayindex(array_index), 4).Value
        update_index = update_index + 1
        array_index = array_index + 1
    End If
Next j

Range(scheduleTable).Cells(1, 1).Select
    newAddress = Range(Selection, Selection.Offset(Z * array_length2 - 1, 5)).address
    With ActiveWorkbook.Names(scheduleTable)
            .RefersTo = "='" & worksheetName & "'!" & newAddress
    End With
    
'rep_index = (end_date - start_date + 1) / 7
array_index = 0
update_index = 16

For k = 1 To x
'Do While x > 0 'And ActiveCell.Offset(0, -1).Value <> ""
    If array_index >= array_length Then
        x = x - 1
        array_index = 0
        start_date = start_date + 7
        
    Else
        Cells(update_index, 3) = start_date + dayindex(array_index) - 7 + 1
        update_index = update_index + 1
        array_index = array_index + 1
    End If
Next k

numRow = Range(scheduleTable).Rows.Count
numCol = Range(scheduleTable).Columns.Count


Range(repDay).Cells(1, 1).Select
    newAddress = Range(Selection, Selection.Offset(numRow - 1, 0)).address
    
    With ActiveWorkbook.Names(repDay)
            .RefersTo = "='" & worksheetName & "'!" & newAddress
    End With

Range(repDate).Cells(1, 1).Select
    newAddress = Range(Selection, Selection.Offset(numRow - 1, 0)).address
    
    With ActiveWorkbook.Names(repDate)
            .RefersTo = "='" & worksheetName & "'!" & newAddress
    End With

Range(timeColumn).Cells(1, 1).Select
    newAddress = Range(Selection, Selection.Offset(numRow - 1, 0)).address
    Selection.Formula = timeFormula
        
    If (Range(newAddress).Rows.Count > 1) Then
        Selection.AutoFill Destination:=Range(newAddress)
    End If
    
    With ActiveWorkbook.Names(timeColumn)
            .RefersTo = "='" & worksheetName & "'!" & newAddress
    End With

Range(timeDateColumn).Cells(1, 1).Select
    newAddress = Range(Selection, Selection.Offset(numRow - 1, 0)).address
    Selection.Formula = timeDateFormula
        
    If (Range(newAddress).Rows.Count > 1) Then
        Selection.AutoFill Destination:=Range(newAddress)
    End If
    
    With ActiveWorkbook.Names(timeDateColumn)
            .RefersTo = "='" & worksheetName & "'!" & newAddress
    End With

Range(costColumn).Cells(1, 1).Select
    newAddress = Range(Selection, Selection.Offset(numRow - 1, 0)).address
    Selection.Formula = costFormula
        
    If (Range(newAddress).Rows.Count > 1) Then
        Selection.AutoFill Destination:=Range(newAddress)
    End If
    
    With ActiveWorkbook.Names(costColumn)
            .RefersTo = "='" & worksheetName & "'!" & newAddress
    End With

Range(amountColumn).Cells(1, 1).Select
    newAddress = Range(Selection, Selection.Offset(numRow - 1, 0)).address
    Selection.Formula = amountFormula
        
    If (Range(newAddress).Rows.Count > 1) Then
        Selection.AutoFill Destination:=Range(newAddress)
    End If
    
    With ActiveWorkbook.Names(amountColumn)
            .RefersTo = "='" & worksheetName & "'!" & newAddress
    End With
    
newAddress = Range(scheduleTable).address
    
    If (Range(newAddress).Rows.Count > 1) Then
        Range(scheduleTable).Rows(1).Copy
        Range(newAddress).PasteSpecial Paste:=xlPasteFormats, Operation:=xlNone, _
            SkipBlanks:=False, Transpose:=False
        Application.CutCopyMode = False
    End If


Application.ScreenUpdating = True
Application.Calculation = xlAutomatic

End Sub
```
